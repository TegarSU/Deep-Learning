```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import numpy as np
import re
```

```python
data1 = pd.read_csv("pagination.csv")
data2 = pd.read_csv("pagination2.csv")
data3 = pd.read_csv("pagination3.csv")

frame = [data1,data2,data3]
result = pd.concat(frame)
# result.head()
result = result.dropna(subset=['keluhan','tipe'])
result = result.drop(columns=['pengadu','web-scraper-order','web-scraper-start-url'],axis=1)
result = result[result['keluhan'].map(len) <300]
result.shape
```

```python
# result.head()
```

```python
def normalize(text):

    pattern = re.compile(r"[\n]")
    patterns = re.compile(r'\s+')
    sentence = re.sub(pattern, ' ', text)
    sentence = re.sub(patterns, ' ', sentence)

    return sentence
```

```python
result['keluhan'] = result['keluhan'].apply(normalize)
result.head()
```
